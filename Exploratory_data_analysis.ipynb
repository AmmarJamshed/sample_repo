```python
# create a function for an advanced line plot
import matplotlib
from matplotlib import pyplot as plt

fig,ax = plt.subplots()
ax.plot(df['Tax 5%'], df['gross income'],ls=':',lw=3)
ax.set_xlabel('tax')
ax.set_ylabel('gross income earnt');
```

```python
# how to create 2 comparable plots at the same time
fig, axes = plt.subplots(ncols=2)
df.plot.scatter(x='Tax 5%', y='gross income', ax=axes[0])
df.plot.scatter(x='Quantity', y='gross income', ax=axes[1])
fig.tight_layout
```

 Taxes amount earnt with quantity have a direct relationship with gross_income earnt

```python
# histogramfor age, with 30 int conversions
df['Quantity'].plot.hist(bins=30);
```

10 is the highest labelled quantity

```python
# a function to create two histogram and concatenate them into one
fig, ax = plt.subplots()
ax.hist(df['Quantity'], bins=20, histtype='step', label='Tax 5%')
ax.hist(df['Total'], bins=20, histtype='stepfilled', label='Tax 5%', alpha=0.3, edgecolor='w')
ax.legend()
```

```python
# plotthe boxplot
df[['gross income',	'Tax 5%']].plot.box();
```

The gross income almost equates the tax being paid at a 5% rates
